Now the mean values for the quantitative variable (age and hours played) in the players.csv data set will be calculated

```r
players_mean<-select(players, Age, played_hours)|> 
  summarize(
    mean_age = round(mean(Age, na.rm = TRUE), 2),
    mean_played_hours = round(mean(played_hours, na.rm = TRUE), 2)
  )
players_mean
```
